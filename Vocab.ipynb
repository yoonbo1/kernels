### 1. Sampling bias
is a bias in which a sample is collected in such a way that some members of the intended population have a lower sampling probability than others. It results in a biased sample, a non-random sample <br><br>


### 2. Stratified Sampling
is a method of sampling from a population which can be partitioned into subpopulations. <br><br>
A stratified sample is one that ensures that subgroups (strata) of a given population are each adequately represented within the whole sample population of a research study. For example, one might divide a sample of adults into subgroups by age, like 18–29, 30–39, 40–49, 50–59, and 60 and above <br><br>


### 3. Stratified Shuffle Split
ensures that the train and test sets have approximately the same percentage(ratio) of samples of each target class as the complete set. <br><br>

### 4. Estimators
<b>Estimators.</b> Any object that can estimate some paramters based on a dataset is called an estimator. The estimation itself is performed by the fit() method. It takes only one dataset as a paramter( or two for supervised learning algos; the second dataset contains the labels). Any other parameter would be hyperparamter(imputer's strategy).
<br><br>
<b>google def*</b> An estimator is any object that learns from data; it may be classification, regression or clustering or a transformer that extracts/filters useful features from raw data.
